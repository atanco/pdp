```python
from selenium.common.exceptions import ElementNotInteractableException

def try_to_click(xpath, name="NA"):
    button_list = driver.find_elements_by_xpath(xpath)    
    try:
        button_list[0].click()
        print("Button with name " + name + "was clicked")
    except IndexError:
        print("indexerror")
    except ElementNotInteractableException as e:
        print(e)
```

```python
from  selenium  import  webdriver
import pandas as pd
import lxml

import  time 

asset_urls = ['https://www.investing.com/funds/amundi-msci-wrld-ae-c', 
              'https://www.investing.com/etfs/db-x-trackers-ii-global-sovereign-5', 
              'https://www.investing.com/etfs/ishares-global-corporate-bond-$',
              'https://www.investing.com/etfs/spdr-gold-trust',
              'https://www.investing.com/indices/usdollar'
             ]
              
asset_dfs = []


for asset_url in asset_urls:
    driver = webdriver.Firefox()
    driver.get(asset_url + "-historical-data")


    #print(driver.find_element_by_id("last_last").text)
    time.sleep(1)
    
    signup_close_button_list_xpath = '/html/body/div[6]/div[2]/i'
    
    try_to_click(signup_close_button_list_xpath, "close signup prompt")
    
    try_to_click('//*[@id="onetrust-accept-btn-handler"]', "acceptBtn")

    try_to_click(signup_close_button_list_xpath, "close signup prompt")
    
    # sidebar = driver.find_elements_by_xpath('/html/body/div[5]/section/div[11]')[0] 
    # print(sidebar.get_attribute("classname"))

    # sidebar.click()


    time.sleep(1)
    
    try_to_click(signup_close_button_list_xpath, "close signup prompt")
    
    date_range_widget = driver.find_elements_by_xpath('/html/body/div[5]/section/div[8]/div[3]')[0]
    date_range_widget.click()

    try_to_click(signup_close_button_list_xpath, "close signup prompt")
    
    start_date_input = driver.find_elements_by_id("startDate")[0]
    start_date_input.clear()
    start_date_input.send_keys("01/01/2020")

    end_date_input = driver.find_elements_by_id("endDate")[0]
    end_date_input.clear()
    end_date_input.send_keys("12/31/2020")

    
    
    apply_btn = driver.find_elements_by_id("applyBtn")[0]
    apply_btn.click()

    time.sleep(1)
    try_to_click(signup_close_button_list_xpath, "close signup prompt")

    table = driver.find_elements_by_xpath('//*[@id="curr_table"]')[0].get_attribute('outerHTML')
    
    driver.close ()
    #asset_dfs.append()
    pd.read_html(table)[0].to_csv(asset_filenames[asset_urls.index(asset_url)])
```
